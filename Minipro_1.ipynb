```python
import os
import pathlib
```

```python
from datetime import datetime
def unix2datetime(unix):
  return datetime.utcfromtimestamp(unix).strftime('%Y%m%d_%H%M%S')
```

```python
path = ("/content/drive/My Drive/Dataset/Training/Train_dataset")
root_dir = '.'
for path in pathlib.Path(root_dir).iterdir():
  info = path.stat()
  ctime = info.st_ctime
  mtime = info.st_mtime
  if mtime <= ctime:
    date_created = unix2datetime(mtime)
  else:
    date_created = unix2datetime(ctime)
  file_name, file_extension = os.path.splitext(path)
  if file_extension != 'JPG':
    continue
new_filename = date_created + file_extension
os.rename(path, new_filename)
print(file_name, new_filename)
```
